```
import re

# Implement	the	following	function	for	determining	the	Jaccard	similarity	
# between	two	strings:

def ngrams(string, n):
    string = re.sub(r'[,-.\/]|\s',r'', string)
    ngrams = zip(*[string[i:] for i in range(n)])
    return [''.join(ngram) for ngram in ngrams]

def jaccard(str1, str2):
    str1 = set(str1)
    str2 = set(str2)
    return float(len(str1 & str2)) / len(str1 | str2)
  
def jaccard_sim(s1,s2,k):
    s1, s2 = ngrams(s1, k), ngrams(s2, k)
    result = jaccard(s1, s2)
    print(result)

def test_jaccard(s1 =	"Apple	Corporation,	CA", s2	=	"Apple	Corp,	CA"):
    jaccard_sim(s1, s2, 1)
    jaccard_sim(s1, s2, 2)
    jaccard_sim(s1, s2, 3)
    jaccard_sim(s1, s2, 4)
    
test_jaccard()

#	Implement	the	following	function	for	computing	the	edit	distance	
# between	two	strings:

def edit_distance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2+1]
        for i1, c1 in enumerate(s1):
            if c1 == c2:
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1],
                                           distances_[-1])))
        distances = distances_
    return distances[-1]
  
def test_edit(s1 =	"Apple", s2	=	"Apel"):
    print(edit_distance(s1, s2))
    
test_edit()

# Implement	the	following	function	for	computing	the	similarity	
# between	two	strings	based	on	their	edit	distance	score:

def ed_similarity(x, y):
    return 1 - edit_distance(x, y)/max(len(x), len(y))
  
def test_ed(s1	=	"Apple	Corporation,	CA",	s2	=	"Apple	Corp,	CA"):
    print(ed_similarity(s1, s2))

test_ed()  
```
